```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
post_df_full1=post_df_full
post_df_full1[['time', 'reactions', 'reaction_count']]
```

```python
post_df_full1['reactions'] = post_df_full1['reactions'].apply(lambda x : dict(eval(x)) )
post_df_full_reactions = post_df_full1['reactions'].apply(pd.Series )
```

```python
post_df_full_with_reactions = pd.concat([post_df_full1, post_df_full_reactions], axis=1).drop('reactions', axis=1)
post_df_full_with_reactions[['post_text','th√≠ch','y√™u th√≠ch','haha','wow','bu·ªìn','ph·∫´n n·ªô','th∆∞∆°ng th∆∞∆°ng',
                                 'shares','comments','reaction_count']]
```

```python
#x·ª≠ l√≠ NaN
columns_to_fill = ['th√≠ch', 'y√™u th√≠ch', 'haha', 'wow', 'bu·ªìn', 'ph·∫´n n·ªô', 'th∆∞∆°ng th∆∞∆°ng', 'shares', 'comments']

post_df_full_with_reactions[columns_to_fill] = post_df_full_with_reactions[columns_to_fill].fillna(0)
post_df_full_with_reactions[['post_text','th√≠ch','y√™u th√≠ch','haha','wow','bu·ªìn','ph·∫´n n·ªô','th∆∞∆°ng th∆∞∆°ng',
                                 'shares','comments','reaction_count']]
```

```python
#v·∫Ω bi·ªÉu ƒë·ªì l∆∞·ª£t react theo ng√†y
fig, ax = plt.subplots(figsize=(20, 10))

columns_to_plot = ['reaction_count']
markers = ["o"]
labels = ["Reaction Count"]

for col, marker, label in zip(columns_to_plot, markers, labels):
    ax.plot(post_df_full_with_reactions['time'], post_df_full_with_reactions[col], label=label, marker=marker)

plt.legend()
from matplotlib.dates import DateFormatter
ax.xaxis.set_major_formatter(DateFormatter('%d-%m-%y %H:%M'))
plt.show()
```

```python
# Danh s√°ch c√°c c·ªôt d·ªØ li·ªáu cho t·ª´ng lo·∫°i t∆∞∆°ng t√°c
interaction_columns = ['th√≠ch', 'y√™u th√≠ch', 'haha', 'wow', 'bu·ªìn', 'ph·∫´n n·ªô', 'th∆∞∆°ng th∆∞∆°ng', 'shares', 'comments']

# V·∫Ω m·ªói c·∫∑p lo·∫°i t∆∞∆°ng t√°c trong m·ªôt h√†ng
fig, axes = plt.subplots(nrows=len(interaction_columns)//2, ncols=2, figsize=(15, 20))

for i in range(0, len(interaction_columns), 2):
    ax1 = axes[i//2, 0]
    ax2 = axes[i//2, 1]

    col1 = interaction_columns[i]
    col2 = interaction_columns[i+1] if i+1 < len(interaction_columns) else None

    ax1.bar(post_df_full_with_reactions['time'], post_df_full_with_reactions[col1], label=col1.capitalize(), width=0.3)
    ax1.set_ylabel(col1.capitalize())  # ƒê·∫∑t nh√£n cho lo·∫°i t∆∞∆°ng t√°c ƒë·∫ßu ti√™n
    ax1.legend()
    ax1.set_xticklabels([])  # X√≥a nh√£n tr·ª•c x

    if col2:
        ax2.bar(post_df_full_with_reactions['time'], post_df_full_with_reactions[col2], label=col2.capitalize(), width=0.3)
        ax2.set_ylabel(col2.capitalize())  # ƒê·∫∑t nh√£n cho lo·∫°i t∆∞∆°ng t√°c th·ª© hai
        ax2.legend()
        ax2.set_xticklabels([])  # X√≥a nh√£n tr·ª•c x

plt.tight_layout()
plt.show()
```

### **PH√ÇN T√çCH POST**

```python
# Chuy·ªÉn c·ªôt 'time' sang ƒë·ªãnh d·∫°ng datetime
post_df_full_with_reactions['time'] = pd.to_datetime(post_df_full_with_reactions['time'])

# T·∫°o c·ªôt 'weekday' ƒë·ªÉ x√°c ƒë·ªãnh ng√†y trong tu·∫ßn (Monday=0, Sunday=6)
post_df_full_with_reactions['weekday'] = post_df_full_with_reactions['time'].dt.dayofweek

# T·∫°o dictionary ƒë·ªÉ √°nh x·∫° s·ªë nguy√™n t·ª´ 0 ƒë·∫øn 6 th√†nh t√™n c·ªßa ng√†y trong tu·∫ßn
weekday_names = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}

# ƒê·ªïi s·ªë nguy√™n th√†nh t√™n c·ªßa ng√†y trong tu·∫ßn
post_df_full_with_reactions['weekday'] = post_df_full_with_reactions['weekday'].map(weekday_names)

# Th·ªëng k√™ s·ªë l∆∞·ª£ng b√†i ƒëƒÉng theo t·ª´ng ng√†y trong tu·∫ßn
posts_by_weekday = post_df_full_with_reactions['weekday'].value_counts()

# S·∫Øp x·∫øp l·∫°i theo th·ª© t·ª± c√°c ng√†y trong tu·∫ßn
posts_by_weekday = posts_by_weekday.reindex(list(weekday_names.values()))

# V·∫Ω bi·ªÉu ƒë·ªì c·ªôt th·ªÉ hi·ªán s·ªë l∆∞·ª£ng b√†i ƒëƒÉng trong t·ª´ng ng√†y trong tu·∫ßn
plt.figure(figsize=(10, 6))
posts_by_weekday.plot(kind='bar', color='skyblue')
plt.xlabel('Ng√†y trong tu·∫ßn')
plt.ylabel('S·ªë l∆∞·ª£ng b√†i ƒëƒÉng')
plt.title('S·ªë l∆∞·ª£ng b√†i ƒëƒÉng c·ªßa page trong t·ª´ng ng√†y trong tu·∫ßn')
plt.xticks(rotation=45)
plt.show()
```

```python
#l·∫•y th√¥ng tin gi·ªù t·ª´ c·ªôt 'time'
post_df_full_with_reactions['hour'] = post_df_full_with_reactions['time'].dt.hour

#th·ªëng k√™ s·ªë l∆∞·ª£ng b√†i ƒëƒÉng theo t·ª´ng gi·ªù trong ng√†y
posts_by_hour = post_df_full_with_reactions['hour'].value_counts().sort_index()

plt.figure(figsize=(10,6))
posts_by_hour.plot(kind='bar', color='skyblue')
plt.xlabel('Gi·ªù trong ng√†y')
plt.ylabel('S·ªë l∆∞·ª£ng b√†i ƒëƒÉng')
plt.title('S·ªë l∆∞·ª£ng b√†i ƒëƒÉng c·ªßa page theo t·ª´ng gi·ªù trong ng√†y')
plt.xticks(rotation=0)
plt.show()
```

```python
reaction_by_hour = post_df_full_with_reactions.groupby('hour')['reaction_count'].sum().reset_index()

posts_count_by_hour = post_df_full_with_reactions['hour'].value_counts().reset_index()
posts_count_by_hour.columns = ['hour', 'post_count']

merged_df = pd.merge(reaction_by_hour, posts_count_by_hour, on='hour')

merged_df['avg_reactions_per_post'] = merged_df['reaction_count']/merged_df['post_count']

plt.figure(figsize=(10,6))
sns.barplot(x='hour', y='avg_reactions_per_post', data=merged_df)
plt.xlabel('Gi·ªù trong ng√†y')
plt.ylabel('Trung b√¨nh s·ªë l∆∞·ª£ng ph·∫£n ·ª©ng m·ªói b√†i ƒëƒÉng')
plt.title('Trung b√¨nh s·ªë l∆∞·ª£ng ph·∫£n ·ª©ng m·ªói b√†i ƒëƒÉng theo t·ª´ng khung gi·ªù trong ng√†y')

plt.xticks(np.arange(0, 24, step=1), [str(i) + 'h' for i in range(24)])

plt.show()
```

```python
import matplotlib.pyplot as plt
# ƒê·∫øm s·ªë b√†i vi·∫øt c√≥ video (video_id kh√°c 0)
posts_with_video = len(post_df_full_with_reactions[post_df_full_with_reactions['video_id'] != 0])

# ƒê·∫øm s·ªë b√†i vi·∫øt c√≥ ·∫£nh (image kh√°c 0)
posts_with_image = len(post_df_full_with_reactions[post_df_full_with_reactions['image'] != 0])

# ƒê·∫øm s·ªë b√†i vi·∫øt kh√¥ng c√≥ video v√† kh√¥ng c√≥ ·∫£nh (c·∫£ video_id v√† image ƒë·ªÅu b·∫±ng 0)
posts_without_image_and_video = len(post_df_full_with_reactions[(post_df_full_with_reactions['video_id'] == 0) & (post_df_full_with_reactions['image'] == 0)])
categories = ['C√≥ video', 'C√≥ ·∫£nh', 'Kh√¥ng c√≥ video v√† ·∫£nh']
values = [posts_with_video, posts_with_image, posts_without_image_and_video]

plt.figure(figsize=(8, 8))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.axis('equal')
plt.show()
```

```python
# T√≠nh trung b√¨nh s·ªë l∆∞·ª£ng t∆∞∆°ng t√°c tr√™n m·ªói lo·∫°i b√†i vi·∫øt
avg_interaction_with_image = post_df_full_with_reactions[post_df_full_with_reactions['image'] != 0][['reaction_count', 'comments', 'shares']].mean().sum()
avg_interaction_with_video = post_df_full_with_reactions[post_df_full_with_reactions['video_id'] != 0][['reaction_count', 'comments', 'shares']].mean().sum()
avg_interaction_without_image_and_video = post_df_full_with_reactions[(post_df_full_with_reactions['image'] == 0) & (post_df_full_with_reactions['video_id'] == 0)][['reaction_count', 'comments', 'shares']].mean().sum()

# T·∫°o d·ªØ li·ªáu cho bi·ªÉu ƒë·ªì
categories = ['C√≥ ·∫£nh', 'C√≥ video', 'Kh√¥ng c√≥ ·∫£nh v√† video']
values = [avg_interaction_with_image, avg_interaction_with_video, avg_interaction_without_image_and_video]

plt.figure(figsize=(8, 6))
plt.barh(categories, values, color='skyblue')
plt.xlabel('Trung b√¨nh s·ªë l∆∞·ª£ng t∆∞∆°ng t√°c')
plt.title('So s√°nh trung b√¨nh s·ªë l∆∞·ª£ng t∆∞∆°ng t√°c tr√™n m·ªói lo·∫°i b√†i vi·∫øt')
plt.show()
```
