```python
FANPAGE_LINK ="catengushitpost"
FOLDER_PATH = "/content/drive/MyDrive/Final Project Python/"
COOKIE_PATH = "/content/drive/MyDrive/Final Project Python/cookies.txt"
PAGES_NUMBER = 10
```

```python
from facebook_scraper import get_posts
import pandas as pd
```

```python
from datetime import datetime
import time


first_date = int(datetime(2023, 10, 1).timestamp())
second_date = int(datetime(2023, 11, 1).timestamp())

post_list = []
while True:
    additional_posts = []
    for post in get_posts(FANPAGE_LINK,
                          options={"comments": True, "reactions": True, "allow_extra_requests": True, "share_info": True},
                          extra_info=True, pages=10, cookies=COOKIE_PATH):
        post_time = int(post['time'].timestamp())
        if first_date <= post_time < second_date:
            additional_posts.append(post)
        elif post_time < first_date:
            break

    post_list.extend(additional_posts)

    # Kiểm tra điều kiện dừng
    if len(post_list) >= 100:
        break
    elif not additional_posts:  # Nếu không còn bài viết nào khác để lấy
        break
```

```python
#tạo ra một DataFrame post_df_full chứa thông tin từ tất cả các post trong post_list, mỗi post chuyển thành 1 df sau đó được add vào df lớn là post_df_full
post_df_full = pd.DataFrame(columns = [])
for post in post_list:
    post_entry = post
    fb_post_df = pd.DataFrame.from_dict(post_entry, orient='index')
    fb_post_df = fb_post_df.transpose() #chuyển hàng thành cột
    post_df_full = pd.concat([post_df_full, fb_post_df], ignore_index=True)
    print(post['post_id']+' get')
```
